```python
import matplotlib.pyplot as plt
import pandas as pd
from sqlalchemy import create_engine
import numpy as np

database_path = ''
# postgresql://postgresusername:password@localhost:5432/db_name
engine = create_engine('postgresql://postgres:postgres@localhost:5432/employee_db')
connection = engine.connect()
```

```python
# Query 
salary_data = engine.execute(
"""
SELECT salary, COUNT(salary)
FROM salaries s
GROUP BY s.salary
ORDER BY s.salary ASC;
""")
```

```python
salaries = []

for salary in salary_data:
    salaries.append(salary)
    
# salaries
```

```python
# create a dataframe from the salaries and frequency of those salaries
df = pd.DataFrame({
    'Salary': [salaries[i][0] for i in range(len(salaries))],
    'Count': [salaries[i][1] for i in range(len(salaries))]
})
df.head()
```

```python
# Create a histogram to visualize the most common salary ranges for employees.
df.hist(column='Salary', bins=100, figsize=(20, 10))
plt.title('Histogram of Salary Ranges')
plt.ylabel('Frequency')
plt.xlabel('Salaries')
plt.show()
```

```python
# Create a bar chart of average salary by title.
```

```python
average_salary_by_title = engine.execute(
"""
SELECT ROUND(AVG(salaries.salary), 2), titles.title
FROM salaries
JOIN employees ON employees.emp_no = salaries.emp_no
JOIN titles ON titles.title_id = employees.emp_title_id
GROUP BY titles.title
""")
```

```python
average_salaries = [avg_salary for avg_salary in average_salary_by_title]
```

```python
average_salaries
```

```python
avg_salaries_df = pd.DataFrame({
    'Average Salary': [salary[0] for salary in average_salaries],
    'Title': [salary[1] for salary in average_salaries]
})
avg_salaries_df.head()
```

```python
titles = [salary[1] for salary in average_salaries]
x_range = np.arange(len(titles))


plt.figure(figsize=(20,15))
plt.bar(x_range, avg_salaries_df['Average Salary'], align='center', alpha=0.5, color='red')
plt.xticks(x_range, titles)
plt.ylabel('Average Salary')
plt.xlabel('Title')
plt.title('Average Salary by Title')
plt.show()
```
